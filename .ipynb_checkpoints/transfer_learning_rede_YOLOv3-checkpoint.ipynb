### Transfer Learning

```python
#Criação dos datasets de imagens com as duas novas classes para treino da rede yolov3
!mkdir ../data
!mkdir ../data/lion
!mkdir ../data/eagle
#!mkdir data/train_images
#!rm -f -R /content/data/
```

```python
def generate_train_labels_file():

  train_imgs = []

  #Generating labels for EAGLE
  for file in os.listdir('/content/data/eagle'):
    if file.endswith('json'):
      with open(f'/content/data/eagle/{file}', 'r') as jfile:
        dict_j = json.load(jfile)

        if len(dict_j) < 2:
          train_imgs.append(dict(
              x=dict_j[0]['rectMask']['xMin'],
              y=dict_j[0]['rectMask']['yMin'],
              width=dict_j[0]['rectMask']['width'],
              height=dict_j[0]['rectMask']['height'],
              class_name='eagle',
              img_path=f'/content/data/eagle/{file.replace(".json", ".jpg")}'
              )
          )
        else:
          for i in range(0, len(dict_j)):
            train_imgs.append(dict(
              x=dict_j[i]['rectMask']['xMin'],
              y=dict_j[i]['rectMask']['yMin'],
              width=dict_j[i]['rectMask']['width'],
              height=dict_j[i]['rectMask']['height'],
              class_name='eagle',
              img_path=f'/content/data/eagle/{file.replace(".json", ".jpg")}'
              )
          )

  #Generating labels for LION
  for file in os.listdir('/content/data/lion'):
    if file.endswith('json'):
      with open(f'/content/data/lion/{file}', 'r') as jfile:
        dict_j = json.load(jfile)

        if len(dict_j) < 2:
          train_imgs.append(dict(
              x=dict_j[0]['rectMask']['xMin'],
              y=dict_j[0]['rectMask']['yMin'],
              width=dict_j[0]['rectMask']['width'],
              height=dict_j[0]['rectMask']['height'],
              class_name='lion',
              img_path=f'/content/data/lion/{file.replace(".json", ".jpg")}'
              )
          )
        else:
          for i in range(0, len(dict_j)):
            train_imgs.append(dict(
              x=dict_j[i]['rectMask']['xMin'],
              y=dict_j[i]['rectMask']['yMin'],
              width=dict_j[i]['rectMask']['width'],
              height=dict_j[i]['rectMask']['height'],
              class_name='lion',
              img_path=f'/content/data/lion/{file.replace(".json", ".jpg")}'
              )
          )

  return train_imgs
```

```python
def generate_classes_file():

  with open('/content/data/classes.names', 'w') as file:
    file.write('eagle\n')
    file.write('lion\n')

  return
```

```python
def generate_train_file():

  images_meta = generate_train_labels_file()

  for img in images_meta:

    if img['class_name'] == 'eagle':
      base_file = f'/content/data/eagle/{os.path.basename(img["img_path"]).replace("jpg", "txt")}'
    else:
      base_file = f'/content/data/lion/{os.path.basename(img["img_path"]).replace("jpg", "txt")}'

    if not os.path.exists(base_file):
      with open(base_file, 'w') as file:
        if img['class_name'] == 'eagle':
          file.write(f'0 {img["x"]} {img["y"]} {img["width"]} {img["height"]}\n')
        else:
          file.write(f'1 {img["x"]} {img["y"]} {img["width"]} {img["height"]}\n')
    else:
      with open(base_file, 'a') as file:
          if img['class_name'] == 'eagle':
            file.write(f'0 {img["x"]} {img["y"]} {img["width"]} {img["height"]}\n')
          else:
            file.write(f'1 {img["x"]} {img["y"]} {img["width"]} {img["height"]}\n')

  return
```

```python
def generate_images_meta_train():

  images_meta = generate_train_labels_file()
  images_array = []

  for img in images_meta:
    if 'images' in img['img_path']:
      continue
    else:
      images_array.append(img['img_path'])

  images_array = list(set(images_array))

  with open('/content/data/train.txt', 'w') as file:
    for img in images_array:
      file.write(f'{img}\n')

  return
```

```python
def generating_cfg_file():

  if not os.path.exists('/content/cfg'):
    os.mkdir('/content/cfg')

  with open('/content/cfg/cfg.data', 'w') as file:
    file.write('classes= 2\n')
    file.write('train = /content/data/train.txt\n')
    file.write('names = /content/data/classes.names\n')
    file.write('backup = backup')

  return
```

### Upload the images of Lion and Eagle classes to /content/data/class_name

```python
#### Após o upload das imagens e arquivos .json gerados pelo PixLab (online tool), são gerados os dados de treino

generate_classes_file()
generate_train_file()
generate_images_meta_train()
generating_cfg_file()
```
